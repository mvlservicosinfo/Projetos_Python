```python
import numpy as np
import matplotlib.pyplot as plt
fsin = np.sin(np.linspace(0, 360, 360)*(np.pi/180.))
x = np.linspace(0, 360, 360)
plt.figure().set_figwidth(10,25)
plt.plot(x, fsin)
plt.xlabel("x", fontsize=15)
plt.ylabel("sin(x)",fontsize=15)
plt.title("Funcao Seno")
plt.legend('f(x)=sin(x)')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fsin = ((np.sin(np.linspace(0, 360, 360)*(np.pi/180.)))+2)
x = np.linspace(0, 360, 360)
plt.figure().set_figwidth(15,35)
plt.plot(x, fsin)
plt.xlabel("x", fontsize=15)
plt.ylabel("sin(x)",fontsize=15)
plt.title("Funcao Seno")
plt.legend('f(x)=sin(x)+2')
plt.grid()
plt.show()
```
